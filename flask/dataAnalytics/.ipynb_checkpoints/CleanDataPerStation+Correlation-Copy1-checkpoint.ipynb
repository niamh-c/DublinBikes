```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

# Choose the Station

```python
df = pd.read_csv('DataCleanedAPR4.csv',  keep_default_na=True, sep=',\s+', delimiter=',', index_col=0)
#df = pd.read_csv('DataCleanedMar28th_test.csv',  keep_default_na=True, sep=',\s+', delimiter=',', index_col=0)
#choose location
station=25
df=df.loc[(df['number']==station)]
```

```python

l=list(df.columns.values)
l.remove('last_update')
for feature in l:
    if np.std(df[feature])<.2:
        df=df.drop(feature, 1)
l=list(df.columns.values)
l.remove('last_update')
```

```python
df.shape
```

# CREATE BINS DICT PER STATION

```python
BINS={0:{},1:{},2:{},3:{},4:{},5:{},6:{}}
for day in range(0,7):
    dt=df.loc[(df['day']==day)]
    dt=dt.sort_values(['time'])
    time=0
    for i in dt.index:
        if dt.at[i, 'time'] in BINS[day]:
            df.at[i, 'time']=BINS[day][dt.at[i, 'time']]
        else:
            BINS[day][dt.at[i, 'time']]=float(time)
            df.at[i, 'time']=BINS[day][dt.at[i, 'time']]
            time+=50
```

### Sorting by avail_bikes and then inserting values corresponding to these from a dict.

```python
for i in range(0,7):   
    dt=df.loc[(df['day']==i)]
    dt=dt.reset_index(drop=True)
    continuous_columns = dt[l].columns
    # Calculate correlation of all pairs of continuous features
    corr = dt[continuous_columns].corr()
    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True
    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11,9))
    # Generate a custom colormap - blue and red
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, annot=True, mask=mask, cmap=cmap, vmax=1, vmin=-1,\
                square=True, xticklabels=True, yticklabels=True,linewidths=.5, cbar_kws={"shrink": .5}, ax=ax)
    plt.yticks(rotation = 0)
    plt.xticks(rotation = 90)
    #sns.regplot(x='time', y='avail_bikes', data=df, order =2);
    print()
```

```python
for j in range(0,5):
    dt=df.loc[(df['day']==j)]
    dt=dt.reset_index(drop=True)
    plt.figure(j)
    sns.regplot(x='time', y='avail_bikes', data=dt, order =2).set_title(j);
    print()
```

```python
def clampLB(minV,maxV, Q3, Q1):
    return max(minV, Q1-1.5*(Q3-Q1))
def clampUB(minV, maxV, Q3, Q1):
    return min(maxV, Q3+1.5*(Q3-Q1))
```

```python
dt=df.copy(deep=True);
```

```python
for day in range(0,7):
    df=dt.loc[dt.day==day]
    col='avail_bikes'
    UB=clampUB(df[col].min(), df[col].max(), df[col].quantile(.75), df[col].quantile(.25))
    for i in df.loc[(df[col]>=UB)].index:
        dt.at[i, col]=UB
    LB=clampLB(df[col].min(), df[col].max(), df[col].quantile(.75), df[col].quantile(.25))
    for i in df.loc[(df[col]<=LB)].index:
        dt.at[i, col]=LB
```

```python
for day in range(0,7):
    f = dt.loc[dt.day==day]['avail_bikes'].plot(kind='box', figsize=(5,5)).set_title(day);
    plt.show()
```

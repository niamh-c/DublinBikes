# Choose the Station

```python
dT = pd.read_csv('DataCleanedAPR4.csv',  keep_default_na=True, sep=',\s+', delimiter=',', index_col=0)
for i in dT.index:
    dT.at[i, 'time']=BINS[dT.at[i, 'time']]
file= open("testingnew.txt","w")
file.write(datetime.today().strftime('%Y-%m-%d'))
RushHour=dict()
BINS=pickle.load(open("time_dict","rb"))
for s in dT['number'].unique():
    Models=dict()
    file.write('Station: '+str(s)+'\n')
    df=dT.loc[(dT['number']==s)]
    meandict={0:{},1:{},2:{},3:{},4:{},5:{},6:{}}
    for day in range(0,7):
        dt=df.loc[(df['day']==day)]
        dt=dt.sort_values(['time'])
        mean=list(round(dt.groupby('time')['avail_bikes'].mean()))
        time=list(dt.time.unique())
        for i in range(0, len(mean)):
            meandict[day][time[i]]=mean[i]
    RushHour[s]=dict()
    df=df.reset_index(drop=True)
    stands=df.bike_stands[0]
    df['rush_hr']=0 ##new
    for day in range(0,7):
        for i in df.loc[df['day']==day].index:
            RushHour[s][day]=dict()
            if meandict[day][df.at[i, 'time']]<round(.2*stands):
                df.at[i,'rush_hr']=1
                RushHour[s][day][df.at[i, 'time']]=df.at[i,'rush_hr']
    l=list(df.columns.values)
    l.remove('last_update')
    l.remove('weekday')
    l.remove('rush_hr')
    weekday=df.loc[df.weekday==1]
    weekend=df.loc[df.weekday==0]
    l=list(weekday.columns.values)
    l.remove('last_update')
    for feature in l:
        if np.std(weekday[feature])<.2:
            weekday=weekday.drop(feature, 1)
        if np.std(weekend[feature])<.2:
            weekend=weekend.drop(feature, 1)
    models=[weekday, weekend]
    modelsStr=['weekday', 'weekend']
    count=0
    Models[s]={'weekday':None, 'weekend':None}
    for m in models:
        file.write('Model '+modelsStr[count]+'\n')
        m=m.sort_values(['last_update'])
        m=m.reset_index(drop=True)
        m=m.drop(['last_update'] ,1)
        cont_features = list(m.columns.values)
        cont_features.remove('avail_bikes')
        X = m[cont_features]
        y = m.avail_bikes
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=False)
        X_test=X_test.reset_index(drop=True)
        y_test=y_test.reset_index(drop=True)
        scaler = MinMaxScaler(feature_range=(0, 1));
        X_train = scaler.fit_transform(X_train);
        X_test = scaler.fit_transform(X_test);
        X = scaler.fit_transform(X);
        model= RandomForestRegressor(n_estimators=100, random_state=0)
        model.fit(X_train, y_train);
        rfr_predictions = model.predict(X_test)
        file.write("Features and coeficients:"+str(list(zip(cont_features, model.feature_importances_))))
        scoring = {'abs_error': 'neg_mean_absolute_error', 'squared_error': 'neg_mean_squared_error', 'r2':'r2'}

        scores = cross_validate(model, X_train, y_train, cv=10, scoring=scoring, return_train_score=True)
        file.write("MAE : "+str(abs(scores['test_abs_error'].mean()))+ "| RMSE : "+ str(math.sqrt(abs(scores['test_squared_error'].mean())))+"| r2 : "+str(scores['test_r2'].mean()))
        file.write('\n')
        Models[modelsStr[count]]=model
        Models[modelsStr[count]+'Features']=list(cont_features)
        count+=1
        outfile = open(str(s),'wb')
        pickle.dump(Models,outfile)
        outfile.close()
    file.write('\n')
    file.write('\n')
file.close()
outfile = open('rush_hr_dicts','wb')
pickle.dump(RushHour,outfile)
outfile.close()
```
